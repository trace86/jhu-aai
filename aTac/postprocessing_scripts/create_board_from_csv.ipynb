```python
import pandas as pd
import numpy as np
import ast
```

```python
filename = "./5x5_experiments/gameplay_5x5_experiments/5x5game_p1minimax1_p2minimax1_2022_04_21-03_28_40_PM.csv"

df = pd.read_csv(filename, header=None)
df.columns = ["Previous State", "Next State", "Player", "Cyber Move", "Game ID"]
```

```python
def printBoard(board):
    for i in range(len(board)):
        for j in range(len(board[i])):
            mark = ' '
            if board[i][j] == 1:
                mark = 'X'
            elif board[i][j] == 2:
                mark = 'O'
            if (j == len(board[i]) - 1):
                print(mark)
            else:
                print(str(mark) + "|", end='')
        if (i < len(board) - 1):
            if len(board) == 3:
                print("-----")
            else:
                print("----------")
```

```python
def printBoard_from_history(df, game_id):
    game_df = df[df["Game ID"] == game_id]
    game_list = game_df["Next State"].to_list()
    player = game_df["Player"].to_list()
    cyber = game_df["Cyber Move"].to_list()

    print(f"Game ID: {game_id}")
    
    for i in range(len(game_list)):
        print("Player {0} move".format(player[i]))
        
        cyber_move = ast.literal_eval(cyber[i])
        
        if cyber_move:
            print("Cyber move: {0}".format(cyber_move[0]))
        else:
            print("Chaos Agent Activated")
        
        
        printBoard(ast.literal_eval(game_list[i]))
        print("")
```

```python
# can probably add funcitonality to write to text file, but for now i've just been copy/pasting this output to a text file

for each in np.unique(df["Game ID"]):
    printBoard_from_history(df, game_id=each)
    print("-"*50)
    print("")
```
